```python
%matplotlib inline
%config InlineBackend.figure_format = "retina"

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import astropy.units as u

# Some plotting config
sns.set(
    context="notebook",
    style="darkgrid",
    font="Times New Roman",
    font_scale=1.75,
)

plt.rcParams["figure.figsize"] = (10, 8)
plt.rcParams["mathtext.fontset"] = "stix"
plt.rcParams.update({"text.usetex": True})


import fitter
import fitter.visualize as viz
```

```python
BH_M_highbin = np.load("./47_tuc_highbin/BH_masses.npy")
BH_N_highbin = np.load("./47_tuc_highbin/N_BHs.npy")

BH_M_lowbin = np.load("./47_tuc_lowbin/BH_masses.npy")
BH_N_lowbin = np.load("./47_tuc_lowbin/N_BHs.npy")

BH_M_nobin = np.load("./47_tuc_nobin/BH_masses.npy")
BH_N_nobin = np.load("./47_tuc_nobin/N_BHs.npy")
```

```python
# BH Plot
fig, axes = plt.subplots(1, 2, figsize=(18, 9))

plt.sca(axes[0])

sns.kdeplot(x=BH_M_highbin, label=r"$f_b = 10\%$", lw=2, fill=True, clip=(0, 5000))
sns.kdeplot(x=BH_M_lowbin, label=r"$f_b = 2\%$", lw=2, fill=True, clip=(0, 5000))
sns.kdeplot(x=BH_M_nobin, label=r"$f_b = 0\%$", lw=2, fill=True, clip=(0, 5000))

plt.gca().set_xlim(left=0)


plt.xlabel(r"Mass in BHs $[\mathrm{M}_\odot]$")
plt.ylabel("Probability density")


plt.legend()
plt.sca(axes[1])

sns.kdeplot(x=BH_N_highbin, label=r"$f_b = 10\%$", lw=2, fill=True, clip=(0, 5000))
sns.kdeplot(x=BH_N_lowbin, label=r"$f_b = 2\%$", lw=2, fill=True, clip=(0, 5000))
sns.kdeplot(x=BH_N_nobin, label=r"$f_b = 0\%$", lw=2, fill=True, clip=(0, 5000))


plt.gca().set_xlim(left=0)

plt.xlabel(r"Number of BHs")
plt.ylabel("Probability density")
fig.tight_layout()

plt.legend()

fig.savefig("BH_KDEs.png", bbox_inches="tight", dpi=200)


print()
```
